# Parsing Dates

```python
import datetime
def str_to_datetime(s):
  split = s.split('-')
  year, month, day = int(split[0]), int(split[1]), int(split[2])
  return pd.to_datetime(f"{year}-{month}-{day}")

datetime_object = str_to_datetime('1986-03-19')
datetime_object
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%Y %m %d')
```
